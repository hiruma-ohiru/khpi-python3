### Task

1. Using LaTex, type the equations in the Jupyter notebook cell.
$$
\left\{
    \begin{array}\\
        8.1x_1 + 8.5x_2 = 72 \\
        -3.2x_1 + 15.2x_2 = 42 \\
        10.2x_2 + 8.8x_2 = 205 \\
    \end{array}
\right.
$$


```python
#importing libraries
from matplotlib import pyplot as plt
import np as numpy
```

```python
def x2_1(x1):
    return (72 - 8.1 * x1) / 8.5

def x2_2(x1):
    return (42 + 3.2 * x1) / 15.2

def x2_3(x1):
    return (205 - 10.2 * x1) / 8.8
```

2. Plot equation lines on separate axes, placing them in the same row meeting the following requirements:

```python
fig, ax = plt.subplots(1, 3, figsize=(25,8), sharey=True)

x1 = np.linspace(0, 50, 100)
ax[0].plot(x1, x2_1(x1), color='firebrick', linestyle='dashed')
ax[1].plot(x1, x2_2(x1), color='purple', linestyle='dashdot')
ax[2].plot(x1, x2_3(x1), color='green', linestyle='dotted')

ax[0].set(title='8.1x1+8.5x2=72', xlabel='x1', ylabel ='x2')
ax[1].set(title='−3.2x1+15.2x2=42', xlabel='x1', ylabel ='x2')
ax[2].set(title='10.2x2+8.8x2=205', xlabel='x1', ylabel ='x2')

plt.show()
```

4.Using the submodule numpy.linalg find the point of intersection of all pairs of lines, mark them and make the label on the plot.

```python
A12 = np.array([[8.1, 8.5], [-3.2, 15.2]])
b12 = np.array([72, 42])
x12 = np.linalg.solve(A12, b12)

A13 = np.array([[8.1, 8.5], [10.2, 8.8]])
b13 = np.array([72, 205])
x13 = np.linalg.solve(A13, b13)

A23 = np.array([[-3.2, 15.2], [10.2, 8.8]])
b23 = np.array([42, 205])
x23 = np.linalg.solve(A23, b23)
```

3. Place all lines in one figure, meeting the following requirements

```python
x1 = np.linspace(0, 100, 100)

plt.plot(x1, x2_1(x1), color='firebrick', linestyle='dashed', label='8.1x1+8.5x2=72')
plt.plot(x1, x2_2(x1), color='purple', linestyle='dashdot', label='−3.2x1+15.2x2=42')
plt.plot(x1, x2_3(x1), color='green', linestyle='dotted', label='10.2x2+8.8x2=205')

plt.text(x1[20]-11, x2_1(x1)[20]-11, f'8.1x1+8.5x2=72', rotation=-29)
plt.text(x1[40]+3,  x2_2(x1)[40]+3, f'−3.2x1+15.2x2=42', rotation=7)
plt.text(x1[30]-8,  x2_3(x1)[30]-8, f'10.2x2+8.8x2=205', rotation=-34)

plt.scatter(x12[0], x12[1] , color="red")
plt.scatter(x13[0], x13[1] , color="red")
plt.scatter(x23[0], x23[1] , color="red")

plt.title("Intersection of three lines")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(color='linen', linewidth=2)
plt.figure(figsize=(18,6))

plt.show()
```
